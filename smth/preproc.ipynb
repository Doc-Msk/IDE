```python
import pandas as pd
import requests

# URL файла на Dropbox
url = 'https://www.dropbox.com/s/64ol9q9ssggz6f1/data_ford_price.xlsx?dl=1'

# Скачать файл и сохранить его на диск
response = requests.get(url)
with open('data_ford_price.xlsx', 'wb') as file:
    file.write(response.content)

# Чтение данных в pandas DataFrame с указанием движка openpyxl
df = pd.read_excel('data_ford_price.xlsx', engine='openpyxl')

# Отображение первых строк данных
display(df.head())
```

```python
columns_to_change = ['cylinders', 'title_status',
                     'transmission', 'drive', 'size']

for i in columns_to_change:
    print(df[i].value_counts())
    print()
```

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np
 
one_hot_encoder = OneHotEncoder()
 
# 'учим' и сразу применяем преобразование к выборке, результат переводим в массив
data_onehot = one_hot_encoder.fit_transform(df[columns_to_change]).toarray()
 
# запишем полученные названия новых колонок в отдельную переменную
column_names = one_hot_encoder.get_feature_names_out(columns_to_change)
print(column_names)

data_onehot = pd.DataFrame(data_onehot, index=df.index, columns=column_names)

display(data_onehot)
```

```python
data_new = pd.concat([df, data_onehot], axis=1)

data_new = data_new.drop(columns_to_change, axis=1)
display(data_new)
```
